```python
def genotype_converter (genotype):
    x = 1000
    if genotype == 'A1 B1 / A1 B1':
        x = 0
    elif genotype == 'A1 B2 / A1 B2':
        x = 1
    elif genotype == 'A2 B1 / A2 B1':
        x = 2
    elif genotype == 'A2 B2 / A2 B2':
        x = 3
    elif genotype == 'A1 B1 / A1 B2':
        x = 4
    elif genotype == 'A1 B2 / A1 B1':
        x = 4
    elif genotype == 'A1 B1 / A2 B1':
        x = 5
    elif genotype == 'A2 B1 / A1 B1':
        x = 5
    elif genotype == 'A1 B1 / A2 B2':
        x = 6
    elif genotype == 'A2 B2 / A1 B1':
        x = 6
    elif genotype == 'A1 B2 / A2 B1':
        x = 7
    elif genotype == 'A2 B1 / A1 B2':
        x = 7
    elif genotype == 'A1 B2 / A2 B2':
        x = 8
    elif genotype == 'A2 B2 / A1 B2':
        x = 8
    elif genotype == 'A2 B1 / A2 B2':
        x = 9
    elif genotype == 'A2 B2 / A2 B1':
        x = 9
    return x
```

```python
def death_function (population, individual_index):
    a = population.to_numpy()
    a = np.delete(a, individual_index, axis=0)
    a[0:, 0]  = np.array(range(0, len(a)))
    col1 = "Index"
    col2 = "Genotype"
    col3 = "Mating type"
    col4 = "Parasite status"
    population = pd.DataFrame({col1:a[0:, 0] , col2: a[0:, 1], col3: a[0:, 2], col4: a[0:, 3]})
    return population
```

```python
def infection_function (population, individual_index):
    if population.at[individual_index, 'Parasite status'] == 'infected':
        population.at[individual_index , 'Parasite status'] = 'uninfected'
    if population.at[individual_index, 'Parasite status'] == 'uninfected':
        population.at[individual_index , 'Parasite status'] = 'infected'
    return population
```
